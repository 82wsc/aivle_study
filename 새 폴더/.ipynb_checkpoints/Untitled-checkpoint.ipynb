## BMI

```python
df_body = df[['가입자일련번호','성별코드','연령대코드(5세단위)','신장(5cm단위)','체중(5kg단위)','허리둘레']]
```

```python
df_body.head()
```

```python
df_body['BMI']=df_body['체중(5kg단위)']/(df_body['신장(5cm단위)']/100)
```

```python
df_body.head()
```

```python
bins = [-float('inf'), 18.5, 25, 30, float('inf')]
labels = ['저체중', '정상 체중', '과체중', '비만']
```

```python
df_body['BMI'] = pd.cut(df_body['BMI'], bins=bins, labels=labels)
```

## 혈압

```python
df_bp = df[['가입자일련번호','수축기혈압','이완기혈압']]
```

```python
# 혈압 평가를 위한 조건과 선택지 정의
conditions = [
    (df_bp['수축기혈압'] < 120) & (df_bp['이완기혈압'] < 80),
    (df_bp['수축기혈압'].between(120, 139)) & (df_bp['이완기혈압'].between(80, 89)),
    (df_bp['수축기혈압'].between(140, 159)) & (df_bp['이완기혈압'].between(90, 99)),
    (df_bp['수축기혈압'] >= 160) | (df_bp['이완기혈압'] >= 100),
    (df_bp['수축기혈압'] >= 180) | (df_bp['이완기혈압'] >= 120)
]
```

```python
choices = [
    '정상 혈압',
    '주의 혈압 (고혈압 전단계)',
    '고혈압 1단계',
    '고혈압 2단계',
    '위험 수치'
]
```

```python
# np.select를 사용하여 혈압 평가 열 추가
df_bp['혈압 평가'] = pd.Series(np.select(conditions, choices, default='알 수 없음'))
```

## 콜레스테롤 수치

```python
df_cl = df[['가입자일련번호','총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤']]
```

```python
df_cl.head()
```

```python
# 총콜레스테롤 평가 추가
chol_bins = [-float('inf'), 200, 239, float('inf')]
chol_labels = ['정상 수치', '경계 수치', '높은 수치 (위험)']
df_cl['총콜레스테롤 평가'] = pd.cut(df_cl['총콜레스테롤'], bins=chol_bins, labels=chol_labels)
```

```python
# LDL 콜레스테롤 평가 추가
ldl_bins = [-float('inf'), 100, 129, 159, float('inf')]
ldl_labels = ['정상 수치', '경계 수치', '높은 수치', '매우 높은 수치']
df_cl['LDL 평가'] = pd.cut(df_cl['LDL콜레스테롤'], bins=ldl_bins, labels=ldl_labels)
```

```python
# HDL 콜레스테롤 평가 추가 (남성 기준)
hdl_bins = [-float('inf'), 40, float('inf')]
hdl_labels = ['낮은 수치', '정상 수치']
df_cl['HDL 평가'] = pd.cut(df_cl['HDL콜레스테롤'], bins=hdl_bins, labels=hdl_labels)
```

```python
# 트리글리세라이드 평가 추가
trig_bins = [-float('inf'), 150, 199, 499, float('inf')]
trig_labels = ['정상 수치', '경계 수치', '높은 수치 (위험)', '매우 높은 수치']
df_cl['트리글리세라이드 평가'] = pd.cut(df_cl['트리글리세라이드'], bins=trig_bins, labels=trig_labels)
```

```python
df_cl.head()
```

## 간 기능 평가: AST, ALT, 감마지티피

```python
df_liver = df[['가입자일련번호','성별코드','혈청지오티(AST)', '혈청지피티(ALT)', '감마지티피']]
```

```python
df_liver.head()
```

```python
# 혈청지오티(AST) 평가 추가
chol_bins = [5, 40, float('inf')]
chol_labels = ['정상 수치', '높은 수치']
df_liver['혈청지오티(AST) 평가'] = pd.cut(df_liver['혈청지오티(AST)'], bins=chol_bins, labels=chol_labels)
```

```python
# 혈청지피티(ALT) 평가 추가
chol_bins = [7, 56, float('inf')]
chol_labels = ['정상 수치', '높은 수치']
df_liver['혈청지피티(ALT) 평가'] = pd.cut(df_liver['혈청지피티(ALT)'], bins=chol_bins, labels=chol_labels)
```

```python
# 감마지피티(ALT) 평가 추가

male_bins = [10, 72, float('inf')]
male_labels = ['정상 수치', '높은 수치']

female_bins = [6, 43, float('inf')]
female_labels = ['정상 수치', '높은 수치']

# 성별에 따라 감마지티피 평가 추가
df_liver['감마지티피 평가'] = '알 수 없음'  # 기본값 설정

# 남성의 경우
df_liver.loc[df_liver['성별코드'] == 1, '감마지티피 평가'] = pd.cut(
    df_liver.loc[df_liver['성별코드'] == 1, '감마지티피'],
    bins=male_bins,
    labels=male_labels
)

# 여성의 경우
df_liver.loc[df_liver['성별코드'] == 2, '감마지티피 평가'] = pd.cut(
    df_liver.loc[df_liver['성별코드'] == 2, '감마지티피'],
    bins=female_bins,
    labels=female_labels
)
```

```python
df_liver.head()
```

## 신장 기능 평가: 혈청 크레아티닌

```python
df_kidney = df[['가입자일련번호','성별코드','혈청크레아티닌']]
```

```python
# 신장 기능 평가 추가

male_bins = [0.7, 1.4, float('inf')]
male_labels = ['정상 수치', '높은 수치']

female_bins = [0.6, 1.2, float('inf')]
female_labels = ['정상 수치', '높은 수치']

# 성별에 따라 신장 기능 평가 추가
df_kidney['신장 기능 평가'] = '알 수 없음'  # 기본값 설정

# 남성의 경우
df_kidney.loc[df_kidney['성별코드'] == 1, '신장 기능 평가'] = pd.cut(
    df_kidney.loc[df_kidney['성별코드'] == 1, '혈청크레아티닌'],
    bins=male_bins,
    labels=male_labels
)

# 여성의 경우
df_kidney.loc[df_kidney['성별코드'] == 2, '신장 기능 평가'] = pd.cut(
    df_kidney.loc[df_kidney['성별코드'] == 2, '혈청크레아티닌'],
    bins=female_bins,
    labels=female_labels
)
```
